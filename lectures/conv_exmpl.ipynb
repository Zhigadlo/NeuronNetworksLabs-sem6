```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
tst_img= Image.open('monke.jpg')
test_img_ish = cv2.imread('monke.jpg')
test_img = test_img_ish[:, :, :: -1]

kernel_nochange = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
kernel_gaus = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])/9
kernel_sharp = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
kernel_border_x = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
kernel_border_y = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernel_border_xy = np.array([[-1, -1, -1], [-1, 8, -1], [1, 2, 1]])

img_noch = cv2.filter2D(test_img_ish, -1, kernel_nochange)
img_gaus = cv2.filter2D(test_img_ish, -1, kernel_gaus)
img_sharp = cv2.filter2D(test_img_ish, -1, kernel_sharp)
img_border_x = cv2.filter2D(test_img_ish, -1, kernel_border_x)
img_border_y = cv2.filter2D(test_img_ish, -1, kernel_border_y)
img_border_xy = cv2.filter2D(test_img_ish, -1, kernel_border_xy)

fig, axs = plt.subplots(2, 3, figsize = (16, 8))
axs[0][0].imshow(img_noch[:, :, :: -1])
axs[0][1].imshow(img_gaus[:, :, :: -1])
axs[0][2].imshow(img_sharp[:, :, :: -1])
axs[1][0].imshow(img_border_x[:, :, :: -1])
axs[1][1].imshow(img_border_y[:, :, :: -1])
axs[1][2].imshow(img_border_xy[:, :, :: -1])
```
